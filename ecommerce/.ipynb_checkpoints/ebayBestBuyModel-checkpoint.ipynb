```python
import numpy as np
import pandas as pd
import multiprocessing
import time
import pickle
import string
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from scipy.spatial.distance import pdist

dist_metrics = ['euclidean', 'cityblock', 'chebyshev', 'hamming', 'cosine']
exclude = set(string.punctuation)
```

```python
price_frame = pd.read_csv('ebay_best_buy_data/price_table_3.csv', encoding='utf-8')
ebay_frame = pd.read_csv('ebay_best_buy_data/ebay_table_3.csv', encoding='utf-8')
best_buy_frame = pd.read_csv('ebay_best_buy_data/best_buy_table_3.csv', encoding='utf-8')
match_frame = pd.read_csv('ebay_best_buy_data/match_table_3.csv', encoding='utf-8')
```

```python
cheap_best_buy = price_frame.loc[price_frame.best_buy_price < price_frame.ebay_price]
cheap_best_buy['price_difference'] = cheap_best_buy.ebay_price - cheap_best_buy.best_buy_price
np.sum(cheap_best_buy.price_difference)
# cheap_best_buy.sort('price_difference', ascending=False)
```

```python
match_frame.loc[match_frame.match_ind==0].shape[0]*1.0/match_frame.shape[0]
# match_frame.loc[match_frame.match_ind==0]
```

```python
def dists(sPair):
    returnDists = [-1,-1,-1,-1,-1]
    try:
        s1 = sPair[0]
        s2 = sPair[1]
        vecs = CountVectorizer().fit_transform([s1,s2]).todense()
        dists = list(map(lambda x : pdist(vecs, x)[0], dist_metrics))
        returnDists = []
        for dist in dists:
            if np.isnan(dist):
                returnDists.append(-1)
            else:
                returnDists.append(dist)
    except:
        pass
    return returnDists
```

```python
pool = multiprocessing.Pool(3)
distArr = pool.map(dists, zip(match_frame.ebay_name, match_frame.best_buy_name))
distArr
```

```python
data = np.array(distArr)
match_with_dist = pd.concat([match_frame, pd.DataFrame(data, columns=dist_metrics)], axis=1)
match_with_dist.to_csv('ebay_best_buy_data/match_with_dist_3.csv', encoding='utf-8')
match_with_dist
```

```python
def xWordSame(s1, s2, ind):
    return s1.split()[ind]==s2.split()[ind]

def lenInWords(s):
    return len(s.split())

def differentNumber(s1, s2):
    rtn_ind = False
    s1 = ''.join(ch for ch in s1 if ch not in exclude)
    s2 = ''.join(ch for ch in s2 if ch not in exclude)
    s1_numbers = set([int(s) for s in s1.split() if s.isdigit()])
    s2_numbers = set([int(s) for s in s2.split() if s.isdigit()])
    number_intersection = s1_numbers.intersection(s2_numbers)
    if len(s1_numbers) != len(number_intersection):
        rtn_ind = True
    return rtn_ind

def differentDate(s1, s2):
    rtn_ind = False
    s1 = ''.join(ch for ch in s1 if ch not in exclude)
    s2 = ''.join(ch for ch in s2 if ch not in exclude)
    s1_numbers = set(list(filter(lambda x : len(str(x))==4,[int(s) for s in s1.split() if s.isdigit()])))
    s2_numbers = set(list(filter(lambda x : len(str(x))==4,[int(s) for s in s2.split() if s.isdigit()])))
    number_intersection = s1_numbers.intersection(s2_numbers)
    if len(s1_numbers) != len(number_intersection):
        rtn_ind = True
    return rtn_ind
    
    

newCols = ["lastWordSame","firstWordSame",'bestBuyLen','ebayLen', 'diffNum', 'diffDate']
# match_with_dist["lastWordSame"] = list(map(lambda x: xWordSame(x[0],x[1], -1), zip(match_with_dist.ebay_name, match_with_dist.best_buy_name)))
# match_with_dist["firstWordSame"] = list(map(lambda x: xWordSame(x[0],x[1], 0), zip(match_with_dist.ebay_name, match_with_dist.best_buy_name)))
# match_with_dist['bestBuyLen'] = match_with_dist["best_buy_name"].map(lenInWords)
# match_with_dist['ebayLen'] = match_with_dist["ebay_name"].map(lenInWords)
match_with_dist['diffNum'] = list(map(lambda x: differentNumber(x[0],x[1]), zip(match_with_dist.ebay_name, match_with_dist.best_buy_name)))
match_with_dist['diffDate'] = list(map(lambda x: differentDate(x[0],x[1]), zip(match_with_dist.ebay_name, match_with_dist.best_buy_name)))
match_with_dist[1:10]
```

```python
shuffle = np.random.permutation(data.shape[0])
match_with_dist = match_with_dist.iloc[shuffle]
msk = np.random.rand(data.shape[0]) < .8
train = match_with_dist.iloc[msk]
test = match_with_dist.iloc[~msk]
colsToUse = list(dist_metrics)
for col in newCols:
    colsToUse.append(col)
train_data = train[colsToUse]
train_labels = train['match_ind']
test_data = test[colsToUse]
test_labels = test['match_ind']
```

```python
clf = RandomForestClassifier(100
                             , n_jobs = 3
                             , class_weight={1: 1, 0: 10}
                            )
clf.fit(train_data, train_labels)
preds = clf.predict(test_data)
print("accuracy:\n" + str(np.mean(preds==test_labels)))
wrong = test.loc[preds!=test_labels]
falsePositives = wrong[wrong.match_ind==0]
print("\nfalse positive count:\n" + str(falsePositives.shape[0]))
print("\nfalse positive rate:\n" + str(falsePositives.shape[0] * 1.0 / test.loc[test.match_ind==0].shape[0]))
print("\nfalse positives")
falsePositives
```
